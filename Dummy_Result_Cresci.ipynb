```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
averaged_data = {}
for experiment, metrics in experiment_data.items():
    averaged_data[experiment] = {}
    for metric, values in metrics.items():
        averaged_data[experiment][metric] = np.mean(values, axis=0)


epochs = [10, 20, 30, 40, 50]
metrics_to_plot = ['loss', 'acc', 'auc', 'prec', 'rec', 'f1']
num_cols = 3  # Number of columns per row
num_rows = (len(metrics_to_plot) + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 8), constrained_layout=True)
axes = axes.flatten()

for i, metric in enumerate(metrics_to_plot):
    for experiment, metrics in experiment_data.items():
        averaged_values = np.mean(metrics[metric], axis=0)
        axes[i].plot(epochs, averaged_values, marker='o', label=f'Experiment {experiment}')
    axes[i].set_title(f'{metric.upper()} over Epochs', fontsize=10)
    axes[i].set_xlabel('Epochs', fontsize=8)
    axes[i].set_ylabel(metric.upper(), fontsize=8)
    axes[i].legend(fontsize=8)
    axes[i].grid(True)

# Turn off unused subplots
for j in range(len(metrics_to_plot), len(axes)):
    axes[j].axis('off')

plt.show()
```

```python
results = []
for experiment, metrics in experiment_data.items():
    row = {'Experiment': experiment}
    for metric, values in metrics.items():
        if metric != 'loss':
            avg_epoch_50 = np.mean([seed[4] for seed in values])
            row[metric] = int(10000*avg_epoch_50)/100
    results.append(row)


df = pd.DataFrame(results)
df.set_index('Experiment', inplace=True)


def underline_max(df):
    formatted_rows = []
    for _, row in df.iterrows():
        formatted_row = []
        for col in df.columns:
            value = row[col]
            if value == df[col].max():
                formatted_row.append(f'\033[4m{value:.2f}\033[0m')
            else:
                formatted_row.append(f'{value:.2f}')
        formatted_rows.append(formatted_row)
    return formatted_rows

# Format DataFrame rows with underlining for max values
formatted_rows = underline_max(df)


print("Exp\t" + "\t".join(df.columns))
for experiment, row in zip(df.index, formatted_rows):
    print(f"{experiment}\t" + "\t".join(row))

# # Optional: Plot the table as an image
# plt.figure(figsize=(10, len(results) * 0.6))
# plt.axis('off')
# plt.table(
#     cellText=df.values,
#     colLabels=df.columns,
#     rowLabels=df.index,
#     cellLoc='center',
#     loc='center',
# )
# plt.title('Metrics Averages at Epoch 50', fontsize=14, pad=20)
# plt.show()
```
